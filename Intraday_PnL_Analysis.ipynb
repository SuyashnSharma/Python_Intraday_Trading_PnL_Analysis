# KITE PnL ASSIGNMENT

```python
import numpy as np
import pandas as pd
```

```python
# Load the data
data = pd.read_csv('orders (1).csv')

# Filter only "MIS" products
df = data[data['Product'] == 'MIS']
```

# 1) Different types of charges for Individual trade

```python
# Convert the 'Qty.' column to string type first, then perform the split operation

df['Qty.'] = df['Qty.'].astype(str).str.split('/').str[0].astype(int)
```

```python
#df['Qty.'].astype(str): Converts all the values in the 'Qty.' column to strings.

#This ensures that even if there are non-string values, such as integers or NaN, they are first converted to strings before applying the .str operations.

#.str.split('/'): Now, split the string values by the / character.

#.str[0]: Extracts the first part of the split string 

#.astype(int): Converts the extracted string part to an integer.

df
```

**(A) TURNOVER:-**

```python
df['Turnover'] = df['Qty.'] * df['Avg. price']
df
```

**(B) BROKERAGE-**

```python
df['Brokerage'] = np.where(df['Turnover'] * 0.0003 < 20, df['Turnover'] * 0.0003, 20)
df                                                                            # Brokerage % - 0.03% or Rs. 20/executed order whichever is lower
```

**(C) STT/CTT-**

```python
df["STT/CTT"] = np.where((df["Type"]=="SELL"),df["Turnover"]*0.00025,0) 
df                                                                       # STT/CTT column only for 'SELL' type and set 0 for 'BUY' type
```

**(D) TRANSACTION CHARGES-**

```python
df["Transaction_charges"] = df["Turnover"]* 0.0000297
df                                                        # using 'NSE' VALUES for transaction charges
```

**(E) SEBI-**

```python
df["Sebi"] = df['Turnover']*10/10000000
df                                      #₹10 / crore on 'Turnover'
```

**(F) GST-**

```python
df['GST'] = (df["Brokerage"]+df["Transaction_charges"]+df['Sebi'])*0.18
df
```

**(G) STAMP CHARGES-**

```python
df["Stamp_Charges"] = np.where((df["Type"]=="BUY"),df["Turnover"]*0.00003,0)
df                                                        #'stamp charges' - 0.003% or ₹300 / crore on 'BUY' side and 0 for 'SELL' side
```

**(H) TOTAL CHARGES-**

```python
df['Total_Charges'] = df["Brokerage"]+df['STT/CTT']+df['Transaction_charges']+df['Sebi']+df['GST']+df['Stamp_Charges']
df
```

# 2) Stock wise and Type wise analysis with weighted Avg. price & calculated charges

```python
# Weighted Average Price = sum of (Qty×Avg. Price)/sum of Qty:-
def weighted_avg_price(group):
    total_qty = group['Qty.'].sum()
    weighted_avg = (group['Qty.'] * group['Avg. price']).sum() / total_qty
    return weighted_avg
```

```python
# Group by 'Instrument' and 'Type' and sum other columns
stock_type_wise = df.groupby(['Instrument', 'Type'])[['Qty.','Turnover','Brokerage','STT/CTT','Transaction_charges','Sebi','GST','Stamp_Charges','Total_Charges']].sum()

# Apply the custom function for weighted average price
stock_type_wise['Weighted_Avg_Price'] = df.groupby(['Instrument', 'Type']).apply(weighted_avg_price)

# Reset index for easier view
stock_type_wise = stock_type_wise.reset_index()

# Print the result
stock_type_wise
```

# 3) Overall Summary of each Stocks

```python
# 1. Calculate Gross PnL for each instrument (SELL turnover - BUY turnover)

def calculate_gross_pnl(df):
    sell_turnover = df[df['Type'] == 'SELL']['Turnover'].sum()
    buy_turnover = df[df['Type'] == 'BUY']['Turnover'].sum()
    return sell_turnover - buy_turnover
```

```python
# Applying  the Gross PnL function to the grouped data by Instrument

gross_pnl = df.groupby('Instrument').apply(calculate_gross_pnl).reset_index(name='Gross PnL')
```

```python
# 2. Calculate Total Charges for each instrument

def calculate_total_charges(df):
    return df['Total_Charges'].sum()

total_charges = df.groupby('Instrument').apply(calculate_total_charges).reset_index(name='Total_Charges')
```

```python
# 3. Merge Gross PnL and Total Charges into a summary

summary = pd.merge(gross_pnl, total_charges, on='Instrument')
```

```python
# 4. Calculate Net PnL (Gross PnL - Total Charges)

summary['Net PnL'] = summary['Gross PnL'] - summary['Total_Charges']

# 5. Calculate % Charges on Gross PnL

summary['% Charges on Gross PnL'] = (summary['Total_Charges'] / summary['Gross PnL']) * 100

# converting it to dataframe:

summary_data = pd.DataFrame(summary).round()
```

```python
# Display the final summary
summary_data
```

```python
with pd.ExcelWriter('stock_data.xlsx') as excel:
    df.to_excel(excel,sheet_name='Types of charges',index=False),
    stock_type_wise.to_excel(excel,sheet_name='Stock & Type wise analysis',index=False)
    summary_data.to_excel(excel,sheet_name='Overall Summary',index=False)
```
